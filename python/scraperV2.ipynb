```python
# Import dependencies
from bs4 import BeautifulSoup
import requests
from splinter import Browser 
import time
```

```python
# Setup Chromedriver
executable_path = {'executable_path': 'chromedriver/chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Test Scrape Loop for data from IGDB for Cloud Database
url = 'https://www.igdb.com/discover'
browser.visit(url)
import numpy as np
for i in range(len(titles)):
    data_row = []
    # Search Input
    try:
        browser.find_by_id('search').fill(f"{titles.iloc[i]['title']} \n")
        more_info_link = ''
    except:
        continue
    try:
        first_found = browser.find_by_id('search-results').find_by_css('div.game-result').first
        more_info_link = first_found.find_by_css('a')['href']
        data_row.append(more_info_link)
        browser.visit(first_found.find_by_css('a')['href'])
    except:
        data_row.append(more_info_link)
        continue
    time.sleep(1)
    # title
    title = ''
    try:
        title = browser.find_by_css('h1.banner-title').text
        data_row.append(title)    
    except:
        continue
    # Systems
    systems = []
    try: 
        for block in browser.find_by_css('div.gamepage-tabs').find_by_css('a'):
            if 'platforms' in block['href']:
                systems.append(block.text)
        data_row.append(systems)
    except:
        data_row.append(systems)
    # Genres and Play Mode
    try:
        info_section_1 = browser.find_by_css('div.optimisly-game-extrainfo1')
        genres = []
        playMode = []
        for block in info_section_1.find_by_css('a.block'):
            if block['itemprop'] == 'playMode':
                playMode.append(block.text)
            elif block['itemprop'] == 'genre':
                genres.append(block.text)
        data_row.append(genres)
        data_row.append(playMode)
        # Themes Series and Player Perspectives
        info_section_2 = browser.find_by_css('div.optimisly-game-extrainfo2')
        themes = []
        series = ''
        playerPerspectives = []
        for block in info_section_2.find_by_css('a.block'):
            string = str(block['href'])
            if 'themes' in string:
                themes.append(block.text)
            if 'collections' in string:
                series = block.text
            if 'perspectives' in string:
                playerPerspectives.append(block.text)
        data_row.append(themes)
        data_row.append(series)
        data_row.append(playerPerspectives)
    except:
        continue
    # Image Link
    try:
        image = browser.find_by_css('img.img-responsive.cover_big')['src']
        data_row.append(image)
    except:
        continue
    # Member Rating
    try:
        critic_rating = browser.find_by_css('div.gamepage-gauge').find_by_css('svg').last.text
    except:
        continue
    data_row.append(critic_rating)
    # Game Description
    try:
        browser.find_by_css('div.gamepage-tabs').find_by_css('div')[6].find_by_css('span').click()
        game_description = browser.find_by_css('div.gamepage-tabs').find_by_css('div')[6].text
    except:
        try:
            game_description = browser.find_by_css('div.gamepage-tabs').find_by_css('div')[6].text
        except:
            continue
    if game_description == 'This game is missing a summary, register an account to add one.':
        continue
    game_description = game_description.replace('\n', '').replace('...', '. ')
    data_row.append(game_description)
    # Keywords
    try:
        browser.find_by_css('div.optimisly-game-extrainfo2').find_by_css('div')[3].find_by_css('span').click()
        keywords = browser.find_by_css('div.optimisly-game-extrainfo2').find_by_css('div')[3].text
    except:
        try:
            keywords = browser.find_by_css('div.optimisly-game-extrainfo2').find_by_css('div')[3].text
        except:
            keywords = ''
    keywords = keywords.replace(',', '')
    data_row.append(keywords)
    # Append Row to DataFrame
    if i == 0:
        test_df = pd.DataFrame(np.reshape(data_row, (1,12)), columns = ['title','moreInfo', 'systems', 'genres',
                                                                        'playModes', 'themes', 'series', 'playerPerspectives',
                                                                        'imageLink', 'memberRating', 'gameDescription',
                                                                        'keywords'])
        
    else:
        df_length = len(test_df)
        test_df.loc[df_length] = data_row
test_df
```
