```python
# import dependencies
import pandas as pd
```

```python
data = pd.read_csv("List of Every Video Game Ever.txt", sep ="\t", error_bad_lines=False)
data
```

```python
# Remove Bad Title Formats
formatted = []
accepted_text = ''
for i in range(len(data)):
    text = data.iloc[i]['title']
    if '(Xbox)' in text:
        accepted_text = text
    elif '(X360)' in text:
        accepted_text = text
    elif '(Xbox One)' in text:
        accepted_text = text
    elif '(PS1)' in text:
        accepted_text = text
    elif '(PS2)' in text:
        accepted_text = text
    elif '(PS3)' in text:
        accepted_text = text
    elif '(PS4)' in text:
        accepted_text = text
    elif '(SNES)' in text:
        accepted_text = text
    elif '(Atari 2600)' in text:
        accepted_text = text
    elif '(Mega Drive / Genesis)' in text:
        accepted_text = text
    elif '(GBA)' in text:
        accepted_text = text
    elif '(Game Boy)' in text:
        accepted_text = text
    elif '(GameCube)' in text:
        accepted_text = text
    elif '(Wii)' in text:
        accepted_text = text
    if '•' in accepted_text:
        continue
    if '[' in accepted_text:
        continue
    if '–' in accepted_text:
        continue
    sep = '('
    accepted_text = accepted_text.split(sep, 1)[0]
    if '!' in accepted_text:
        continue
    elif '#' in accepted_text:
        continue
    if '�' in accepted_text:
        continue
    if '?' in accepted_text:
        continue
    if ')' in accepted_text:
        continue
    formatted.append(accepted_text)

titles = pd.DataFrame(formatted).drop_duplicates().reset_index(drop=True).iloc[1:]
titles.columns = ['title']
titles
```

```python
# Import dependencies
from bs4 import BeautifulSoup
import requests
from splinter import Browser 
import time
```

```python
# Setup Chromedriver
executable_path = {'executable_path': 'chromedriver/chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Test Scrape Loop for data from IGDB for Cloud Database
url = 'https://www.igdb.com/discover'
browser.visit(url)
import numpy as np
for i in range(len(titles)):
    data_row = []
    # Search Input
    try:
        browser.find_by_id('search').fill(f"{titles.iloc[i]['title']} \n")
        more_info_link = ''
    except:
        continue
    try:
        first_found = browser.find_by_id('search-results').find_by_css('div.game-result').first
        more_info_link = first_found.find_by_css('a')['href']
        data_row.append(more_info_link)
        browser.visit(first_found.find_by_css('a')['href'])
    except:
        data_row.append(more_info_link)
        continue
    time.sleep(3)
    # title
    title = ''
    try:
        title = browser.find_by_css('h1.banner-title').text[:-4]
        data_row.append(title)    
    except:
        continue
    # Systems
    systems = []
    try: 
        for block in browser.find_by_css('div.gamepage-tabs').find_by_css('a'):
            if 'platforms' in block['href']:
                systems.append(block.text)
        data_row.append(systems)
    except:
        data_row.append(systems)
    # Genres and Play Mode
    try:
        info_section_1 = browser.find_by_css('div.optimisly-game-extrainfo1')
        genres = []
        playMode = []
        for block in info_section_1.find_by_css('a.block'):
            if block['itemprop'] == 'playMode':
                playMode.append(block.text)
            elif block['itemprop'] == 'genre':
                genres.append(block.text)
        data_row.append(genres)
        data_row.append(playMode)
        # Themes Series and Player Perspectives
        info_section_2 = browser.find_by_css('div.optimisly-game-extrainfo2')
        themes = []
        series = ''
        playerPerspectives = []
        for block in info_section_2.find_by_css('a.block'):
            string = str(block['href'])
            if 'themes' in string:
                themes.append(block.text)
            if 'collections' in string:
                series = block.text
            if 'perspectives' in string:
                playerPerspectives.append(block.text)
        data_row.append(themes)
        data_row.append(series)
        data_row.append(playerPerspectives)
    except:
        continue
    # Image Link
    try:
        image = browser.find_by_css('img.img-responsive.cover_big')['src']
        data_row.append(image)
    except:
        continue
    # Member Rating
    try:
        critic_rating = browser.find_by_css('div.gamepage-gauge').find_by_css('svg').last.text
        if critic_rating == 'N/A':
            try:
                critic_rating = browser.find_by_css('div.gamepage-gauge').find_by_css('svg').first.find_by_css('text').first.text
            except:
                continue
    except:
        try:
            critic_rating = browser.find_by_css('div.gamepage-gauge').find_by_css('svg').first.find_by_css('text').first.text
        except:
            continue
    if critic_rating == 'N/A':
        continue
    data_row.append(critic_rating)
    # Game Description
    try:
        browser.find_by_css('div.gamepage-tabs').find_by_css('div')[6].find_by_css('span').click()
        game_description = browser.find_by_css('div.gamepage-tabs').find_by_css('div')[6].text
    except:
        try:
            game_description = browser.find_by_css('div.gamepage-tabs').find_by_css('div')[6].text
        except:
            continue
    if game_description == 'This game is missing a summary, register an account to add one.':
        continue
    if game_description == 'Official Website':
        continue
    if game_description == 'Hmm.  no websites found for this game. Care to add one?':
        continue        
    game_description = game_description.replace('\n', '').replace('...', '. ')
    data_row.append(game_description)
    # Append Row to DataFrame
    if i == 0:
        test_df = pd.DataFrame(np.reshape(data_row, (1,11)), columns = ['moreInfo','title', 'systems', 'genres',
                                                                        'playModes', 'themes', 'series', 'playerPerspectives',
                                                                        'imageLink', 'memberRating', 'gameDescription'])
        
    else:
        df_length = len(test_df)
        test_df.loc[df_length] = data_row
    print(f"{i/len(titles)}% rows have been iterated on...")
test_df
```
